```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 10, 40)
x
```

```python
sinx = np.sin(x)
sinx
```

## Ploting vectors using numpy

```python
pp.plot(x, sinx)
```

```python
cosx = np.cos(x)
```

```python
pp.plot(x, cosx)
```

```python
pp.plot(x, sinx)
pp.plot(x, cosx)
```

```python
tanx = np.tan(x)
pp.plot(x, tanx)
```

```python
pp.plot(x, sinx, 'x')
pp.plot(x, cosx, 'o')
```

```python
y = sinx * cosx
z = cosx**2 - sinx**2

pp.plot(x,y)
pp.plot(x,z)

pp.legend(['sin(x).cos(x)', 'cos(x)^2 - sin(x)^2'])
```

## Dot product of two vectors & outer product

```python
np.dot(cosx, sinx)
```

```python
# Outer product == sinx x cosx_transpose
mat_outer = np.outer(sinx, cosx)
```
